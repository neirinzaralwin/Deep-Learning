## Step to build stock prediction model

* Data Preprocessing
* Building RNN
* Making the prediction and visualization

<code>I use data from this <a href = 'https://finance.yahoo.com/quote/GOOG/history/'>link</a></code>

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

### Data Preprocessing

```python
data = pd.read_csv('GOOG.csv', date_parser = True)
data.head()
# print("The size of the data: {}".format(data.size))
```

```python
data_training = data[data['Date'] < '2019-01-01'].copy()
data_training.head()
```

```python
data_test = data[data['Date'] >= '2019-01-01'].copy()
data_test
```

```python
# Drop columns
training_data = data_training.drop(['Date','Adj Close'], axis = 1)
training_data
```

```python
#Scale the data between 0 and 1
scaler = MinMaxScaler()
training_data = scaler.fit_transform(training_data)
training_data
```

```python
training_data.shape[0]
```

```python
X_train = []
y_train = []

for i in range(60, training_data.shape[0]):
    X_train.append(training_data[i-60: i])
    y_train.append(training_data[i, 0])
```

```python
#convert list to numpy array
X_train, y_train = np.array(X_train), np.array(y_train)
X_train.shape
```

```python
y_train.shape
```

### Building LSTM

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, LSTM, Dropout
```

```python
# regressor for model selection
regressor = Sequential()

# layer1
regressor.add(LSTM(units = 50, activation = 'relu', return_sequences = True, input_shape = (X_train.shape[1], 5)))
regressor.add(Dropout(0.2))
# layer2
regressor.add(LSTM(units = 60, activation = 'relu', return_sequences = True))
regressor.add(Dropout(0.3))
# layer3
regressor.add(LSTM(units = 80, activation = 'relu', return_sequences = True))
regressor.add(Dropout(0.4))
# layer4
regressor.add(LSTM(units = 120, activation = 'relu'))
regressor.add(Dropout(0.5))
# final layer5
regressor.add(Dense(units = 1))
```

```python
regressor.summary()
```

```python
regressor.compile(optimizer = 'adam', loss= 'mean_squared_error')
```

```python
regressor.fit(X_train, y_train, epochs = 10, batch_size = 32)
```

### Testing

<code>To focus on testing, I must recall last 60 days' trained data. </code>

```python
past_60_days = data_training.tail(60)
df = past_60_days.append(data_test , ignore_index= True)
df = df.drop(['Date', 'Adj Close'], axis = 1)
df.head()
```

```python
inputs = scaler.transform(df)
inputs
```

```python
X_test = []
y_test = []

for i in range(60, inputs.shape[0]):
    X_test.append(inputs[i-60: i])
    y_test.append(inputs[i, 0])
```

```python
X_test, y_test  = np.array(X_test), np.array(y_test)
X_test.shape, y_test.shape
```

```python
y_pred = regressor.predict(X_test)
```

```python
y_pred
```

<code> This is scaled value. let's print scaled value! </code> 

```python
# the data are scaled with these values
scaler.scale_
```

```python
# actual value
scale = 1/8.18605127e-04
scale
```

```python
# to get unscaled value
y_pred = y_pred * scale
y_test = y_pred * scale
```

```python
y_pred
```

```python
y_test
```

### Visualization

```python
plt.figure(figsize = (14,5))
plt.plot(y_test, color = 'red', label = 'Real Google Stock Price')
plt.plot(y_pred, color = 'blue', label = "Predicted Google Stock Price")
plt.title('Google Stock Price Prediction')
plt.xlabel('Time')
plt.ylabel('Google Stock Price')
plt.legend()
plt.show()
```
